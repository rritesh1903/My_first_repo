```python
# Do a train-test Split.
from sklearn.model_selection import train_test_split

#
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state = 42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
## Feature Selection, based on Correlation.
X_train.corr()

# correlation btw independent features
# two findings --> Highly +vely correlated or HIghly -vely correlated.
# -ve can be an important feature. But +vely correlated one, we have to try removing it.
```

```python
# Corr plot on independent variables
plt.figure(figsize=(8,6))
corr = X_train.corr()
sns.heatmap(corr, annot = True)
```

```python
corr.columns
```

```python
# Check for multi-collinearity.
def correlation(dataset, threshold):
  col_corr = set()
  corr_matrix = dataset.corr()
  for i in range(len(corr_matrix.columns)):
    for j in range(i):
      if abs(corr_matrix.iloc[i,j]) >= threshold:
        colname = corr_matrix.columns[i]
        col_corr.add(colname)
  return col_corr
```

```python
#calling correlation function:
correlation(X_train,0.90)   # Threshold here to be decided by Domain Expert.

# BUI seems to be Highly multi-collineared. Therefore it can be planned to remove BUI from Aanalysis.
```

```python
#calling correlation function:
correlated_features = correlation(X_train, 0.85)
print(correlated_features)
```

```python
#Drop correlated features.
X_train.drop(correlated_features, axis = 1, inplace = True)
X_test.drop(correlated_features, axis = 1, inplace = True)

print("X_Train Shape: ",X_train.shape)
print("X_test Shape: ",X_test.shape)
```

# Feature Scaling Or Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

#Fit the data:
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)  # --> Always only do transform on Test Data.
```

```python
X_train_scaled
```

# Box plot on STandardized data.
# TO understand teh effect of Standard_Scaler

```python
plt.subplots(figsize = (10,5))
plt.subplot(1,2,1)
sns.boxplot(data = X_train)
plt.title("X_train Before Scaling")
#
plt.subplot(1,2,2)
sns.boxplot(data = X_train_scaled)
plt.title("X_train After Scaling")
```

# Model MAking

```python
# Linear Regression Model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```python
linReg = LinearRegression()
linReg.fit(X_train_scaled, y_train)
y_pred = linReg.predict(X_test_scaled)
#
mae = mean_absolute_error(y_test, y_pred)
#
score = r2_score(y_test, y_pred)

##print
print("Mea Absolute Error: ", mae)
print("R2 Score: ", score)
```

```python
#PLot
plt.subplots(figsize = (12,5))
plt.scatter(y_test, y_pred)
```

1. Above Model looks absolutely fine, wrt Score, MAE & the plot. It shows a linear relationship.
<br>
# Linear Reg Model tends to overfit also,  lets try other methods of Reg to see if any overfitting is taking place

```python
## Apply Ridge & Lasso to avoid Overfitting.

from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```python
lassoReg = Lasso()
lassoReg.fit(X_train_scaled, y_train)
y_pred = lassoReg.predict(X_test_scaled)
#
mae = mean_absolute_error(y_test, y_pred)
#
score = r2_score(y_test, y_pred)

print("LAsso Score")
print("Mea Absolute Error: ", mae)
print("R2 Score: ", score)
#
plt.subplots(figsize = (12,5))
plt.scatter(y_test, y_pred)
```

```python
# Ridge Regression MOdel:
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```python
ridgeReg = Ridge()
ridgeReg.fit(X_train_scaled, y_train)
y_pred = ridgeReg.predict(X_test_scaled)
#
mae = mean_absolute_error(y_test, y_pred)
#
score = r2_score(y_test, y_pred)

print("Ridge Score")
print("Mea Absolute Error: ", mae)
print("R2 Score: ", score)
#
plt.subplots(figsize = (12,5))
plt.scatter(y_test, y_pred)
```
